```
# %% [markdown]
# # Student Performance Analysis and Prediction
#
# This notebook analyzes student data to understand factors influencing academic performance (classified as Low, Medium, High) and trains several machine learning models to predict this performance category.

# %% [code]
# Import necessary libraries
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import time as t
import sklearn.utils as u
import sklearn.preprocessing as pp
import sklearn.tree as tr
import sklearn.ensemble as es
import sklearn.metrics as m
import sklearn.linear_model as lm
import sklearn.neural_network as nn
import numpy as np
#import random as rnd # This was commented out in the original script
import warnings as w
w.filterwarnings('ignore') # Ignore warnings for cleaner output

# %% [markdown]
# ## Load Data
#
# Upload the `Data.csv` file using the cell below.

# %% [code]
# Upload the CSV file to Colab
from google.colab import files
import io

print("Please upload the Data.csv file:")
uploaded = files.upload()

# Check if the file was uploaded correctly
if 'Data.csv' in uploaded:
    # Read the uploaded CSV file into a pandas DataFrame
    data = pd.read_csv(io.BytesIO(uploaded['Data.csv']))
    print("\nData.csv loaded successfully!")
    print("First 5 rows of the dataset:")
    print(data.head())
else:
    print("\nError: Data.csv not found in uploaded files. Please ensure you uploaded the correct file.")
    # Stop execution if file is not loaded
    raise FileNotFoundError("Data.csv not found.")


# %% [markdown]
# ## Exploratory Data Analysis (EDA) - Visualizations
#
# This section provides interactive plotting based on user choice to visualize relationships between student performance ('Class') and various features.

# %% [code]
# Interactive plotting section
ch = 0
while(ch != 10):
    print("\n--- Plotting Menu ---")
    print("1. Marks Class Count Graph")
    print("2. Marks Class Semester-wise Graph")
    print("3. Marks Class Gender-wise Graph")
    print("4. Marks Class Nationality-wise Graph")
    print("5. Marks Class Grade-wise Graph")
    print("6. Marks Class Section-wise Graph")
    print("7. Marks Class Topic-wise Graph")
    print("8. Marks Class Stage-wise Graph")
    print("9. Marks Class Absent Days-wise Graph")
    print("10. No More Graphs (Proceed to Model Training)")
    print("---------------------\n")

    try:
        ch = int(input("Enter Choice (1-10): "))

        if 1 <= ch <= 9:
            print("\nLoading Graph....")
            t.sleep(0.5) # Reduced sleep time for faster interaction in Colab

        if (ch == 1):
            print("\tMarks Class Count Graph")
            plt.figure(figsize=(8, 5)) # Adjust figure size for better display
            axes = sb.countplot(x='Class', data=data, order=['L', 'M', 'H'])
            plt.title('Distribution of Performance Classes')
            plt.xlabel('Performance Class')
            plt.ylabel('Count')
            plt.show()
        elif (ch == 2):
            print("\tMarks Class Semester-wise Graph")
            fig, axesarr = plt.subplots(1, figsize=(10, 6))
            sb.countplot(x='Semester', hue='Class', data=data, hue_order=['L', 'M', 'H'], ax=axesarr) # Use 'ax' instead of 'axes'
            plt.title('Performance Class Distribution by Semester')
            plt.xlabel('Semester')
            plt.ylabel('Count')
            plt.legend(title='Class')
            plt.show()
        elif (ch == 3):
            print("\tMarks Class Gender-wise Graph")
            fig, axesarr = plt.subplots(1, figsize=(10, 6))
            sb.countplot(x='gender', hue='Class', data=data, order=['M', 'F'], hue_order=['L', 'M', 'H'], ax=axesarr)
            plt.title('Performance Class Distribution by Gender')
            plt.xlabel('Gender')
            plt.ylabel('Count')
            plt.legend(title='Class')
            plt.show()
        elif (ch == 4):
            print("\tMarks Class Nationality-wise Graph")
            # Increase figure size for potentially many nationalities
            nationality_count = data['NationalITy'].nunique()
            fig_width = max(10, nationality_count * 0.5) # Adjust width based on number of categories
            fig, axesarr = plt.subplots(1, figsize=(fig_width, 6))
            sb.countplot(x='NationalITy', hue='Class', data=data, hue_order=['L', 'M', 'H'], ax=axesarr)
            plt.title('Performance Class Distribution by Nationality')
            plt.xlabel('Nationality')
            plt.ylabel('Count')
            plt.xticks(rotation=45, ha='right') # Rotate labels if many categories
            plt.legend(title='Class')
            plt.tight_layout() # Adjust layout
            plt.show()
        elif (ch == 5):
            print("\tMarks Class Grade-wise Graph")
            # Ensure GradeID order is correct based on data
            grade_order = sorted(data['GradeID'].unique(), key=lambda x: int(x.split('-')[1]))
            fig, axesarr = plt.subplots(1, figsize=(12, 6)) # Wider figure
            sb.countplot(x='GradeID', hue='Class', data=data, order=grade_order, hue_order = ['L', 'M', 'H'], ax=axesarr)
            plt.title('Performance Class Distribution by Grade ID')
            plt.xlabel('Grade ID')
            plt.ylabel('Count')
            plt.legend(title='Class')
            plt.show()
        elif (ch == 6):
            print("\tMarks Class Section-wise Graph")
            fig, axesarr = plt.subplots(1, figsize=(10, 6))
            sb.countplot(x='SectionID', hue='Class', data=data, hue_order = ['L', 'M', 'H'], ax=axesarr)
            plt.title('Performance Class Distribution by Section ID')
            plt.xlabel('Section ID')
            plt.ylabel('Count')
            plt.legend(title='Class')
            plt.show()
        elif (ch == 7):
            print("\tMarks Class Topic-wise Graph")
            # Increase figure size for potentially many topics
            topic_count = data['Topic'].nunique()
            fig_width = max(10, topic_count * 0.6) # Adjust width based on number of categories
            fig, axesarr = plt.subplots(1, figsize=(fig_width, 6))
            sb.countplot(x='Topic', hue='Class', data=data, hue_order = ['L', 'M', 'H'], ax=axesarr)
            plt.title('Performance Class Distribution by Topic')
            plt.xlabel('Topic')
            plt.ylabel('Count')
            plt.xticks(rotation=45, ha='right') # Rotate labels if many categories
            plt.legend(title='Class')
            plt.tight_layout() # Adjust layout
            plt.show()
        elif (ch == 8):
            print("\tMarks Class Stage-wise Graph")
            fig, axesarr = plt.subplots(1, figsize=(10, 6))
            sb.countplot(x='StageID', hue='Class', data=data, hue_order = ['L', 'M', 'H'], ax=axesarr)
            plt.title('Performance Class Distribution by Stage ID')
            plt.xlabel('Stage ID')
            plt.ylabel('Count')
            plt.legend(title='Class')
            plt.show()
        elif (ch == 9):
            print("\tMarks Class Absent Days-wise Graph")
            fig, axesarr = plt.subplots(1, figsize=(10, 6))
            sb.countplot(x='StudentAbsenceDays', hue='Class', data=data, order=['Under-7', 'Above-7'], hue_order = ['L', 'M', 'H'], ax=axesarr)
            plt.title('Performance Class Distribution by Absence Days')
            plt.xlabel('Student Absence Days')
            plt.ylabel('Count')
            plt.legend(title='Class')
            plt.show()
        elif(ch == 10):
            print("\nExiting plotting section and proceeding to data preprocessing...\n")
            t.sleep(0.5)
        else:
             print("\nInvalid choice. Please enter a number between 1 and 10.")

    except ValueError:
        print("\nInvalid input. Please enter a number.")
    except Exception as e:
        print(f"\nAn error occurred during plotting: {e}")


# %% [markdown]
# ## Data Preprocessing
#
# 1.  **Drop Columns:** Remove features deemed unnecessary for modeling.
# 2.  **Encode Categorical Features:** Convert categorical string features into numerical representations using Label Encoding.
# 3.  **Shuffle Data:** Randomize the order of data rows.
# 4.  **Split Data:** Divide the dataset into training and testing sets.

# %% [code]
# --- Data Preprocessing ---

# Drop columns - keeping only 'raisedhands', 'VisITedResources', 'Discussion', 'StudentAbsenceDays', and the target 'Class'
# Note: The original script dropped many columns, including potentially useful ones like 'gender', 'GradeID' etc.
# It seems the final features used were based on the array creation later:
# arr = np.array([rai, res, dis, absc]) -> This suggests only 4 features were intended for the final prediction part.
# Let's refine the feature selection based on the model training part below which uses:
# feats = data.values[:, 0:4]
# lbls = data.values[:,4]
# This implies the first 4 columns are features and the 5th is the label AFTER preprocessing.
# We need to carefully select columns BEFORE encoding.

# Let's keep the features potentially used in the prediction input section:
# 'raisedhands', 'VisITedResources', 'AnnouncementsView', 'Discussion', 'StudentAbsenceDays'
# And the target: 'Class'
# Also keeping categorical ones needed for encoding before selection:
# 'gender', 'NationalITy', 'PlaceofBirth', 'GradeID', 'SectionID', 'Topic', 'Semester', 'Relation', 'ParentAnsweringSurvey', 'ParentschoolSatisfaction'

# Re-evaluate columns to drop based on the features used in model training and prediction input
# The model training uses `data.values[:, 0:4]` as features and `data.values[:,4]` as labels *after* encoding.
# The prediction input uses `rai`, `res`, `dis`, `absc`.
# Let's assume the intended features are 'raisedhands', 'VisITedResources', 'Discussion', 'StudentAbsenceDays' and the target is 'Class'.

print("Original columns:", data.columns.tolist())

# Select the columns needed for encoding and final features/label
columns_to_keep = [
    'raisedhands', 'VisITedResources', 'Discussion', 'StudentAbsenceDays', # Numerical features potentially used
    'gender', 'NationalITy', 'PlaceofBirth', 'GradeID', 'SectionID', 'Topic', 'Semester', 'Relation',
    'ParentAnsweringSurvey', 'ParentschoolSatisfaction', # Categorical features from original data
    'AnnouncementsView', # Another numerical feature
    'Class' # Target variable
]
# Filter data to keep only necessary columns initially
data_filtered = data[[col for col in columns_to_keep if col in data.columns]].copy() # Ensure columns exist

print("\nColumns kept for processing:", data_filtered.columns.tolist())

# --- Encoding ---
# The original script encoded all object types. Let's replicate that for the filtered data.
print("\nEncoding categorical features...")
encoders = {} # To store encoders if needed later for inverse transform or new data
for column in data_filtered.columns:
    if data_filtered[column].dtype == type(object):
        le = pp.LabelEncoder()
        data_filtered[column] = le.fit_transform(data_filtered[column])
        encoders[column] = le
        print(f"Encoded column: {column}")

print("\nData after encoding (first 5 rows):")
print(data_filtered.head())

# --- Feature Selection ---
# Select the final features and labels based on the original script's apparent intention
# Features: 'raisedhands', 'VisITedResources', 'Discussion', 'StudentAbsenceDays'
# Label: 'Class'
feature_names = ['raisedhands', 'VisITedResources', 'Discussion', 'StudentAbsenceDays']
label_name = 'Class'

# Verify these columns exist after filtering and encoding
if all(f in data_filtered.columns for f in feature_names) and label_name in data_filtered.columns:
    feats = data_filtered[feature_names].values
    lbls = data_filtered[label_name].values
    print(f"\nSelected features: {feature_names}")
    print(f"Selected label: {label_name}")
else:
    missing_cols = [f for f in feature_names if f not in data_filtered.columns]
    if label_name not in data_filtered.columns:
        missing_cols.append(label_name)
    print(f"\nError: Could not find the following required columns after processing: {missing_cols}")
    raise ValueError("Required columns for feature/label selection are missing.")


# --- Shuffle Data ---
# Shuffle features and labels together
feats, lbls = u.shuffle(feats, lbls, random_state=42) # Added random_state for reproducibility
print("\nData shuffled.")

# --- Split Data ---
# Split into training (70%) and testing (30%) sets
split_index = int(len(feats) * 0.70)

feats_Train = feats[:split_index]
feats_Test = feats[split_index:] # Corrected split index
lbls_Train = lbls[:split_index]
lbls_Test = lbls[split_index:] # Corrected split index

print(f"\nData split into:")
print(f"  Training features shape: {feats_Train.shape}")
print(f"  Training labels shape: {lbls_Train.shape}")
print(f"  Testing features shape: {feats_Test.shape}")
print(f"  Testing labels shape: {lbls_Test.shape}")


# %% [markdown]
# ## Model Training and Evaluation
#
# Train and evaluate the following classification models:
# 1.  Decision Tree
# 2.  Random Forest
# 3.  Perceptron
# 4.  Logistic Regression
# 5.  Multi-Layer Perceptron (MLP) Neural Network

# %% [code]
# --- Model Training and Evaluation ---

results = {} # Dictionary to store accuracy results

# --- 1. Decision Tree ---
print("\n--- Training Decision Tree ---")
t.sleep(0.5)
modelD = tr.DecisionTreeClassifier(random_state=42) # Add random_state
modelD.fit(feats_Train, lbls_Train)
lbls_predD = modelD.predict(feats_Test)

print("\nAccuracy measures using Decision Tree:")
# The original script calculated accuracy manually; using metrics.accuracy_score is standard
accD = m.accuracy_score(lbls_Test, lbls_predD)
results['Decision Tree'] = accD
print(m.classification_report(lbls_Test, lbls_predD, target_names=['High', 'Low', 'Medium'])) # Assuming 0:H, 1:L, 2:M based on prediction code later
print(f"\nAccuracy using Decision Tree: {accD:.3f}")

# --- 2. Random Forest ---
print("\n--- Training Random Forest ---")
t.sleep(0.5)
modelR = es.RandomForestClassifier(random_state=42) # Add random_state
modelR.fit(feats_Train, lbls_Train)
lbls_predR = modelR.predict(feats_Test)

print("\nAccuracy Measures for Random Forest Classifier:")
accR = m.accuracy_score(lbls_Test, lbls_predR)
results['Random Forest'] = accR
#print("\nConfusion Matrix: \n", m.confusion_matrix(lbls_Test, lbls_predR)) # Optional: uncomment to see matrix
print(m.classification_report(lbls_Test, lbls_predR, target_names=['High', 'Low', 'Medium']))
print(f"\nAccuracy using Random Forest: {accR:.3f}")

# --- 3. Perceptron ---
print("\n--- Training Perceptron ---")
t.sleep(0.5)
# Standardize features for linear models like Perceptron
scaler = pp.StandardScaler()
feats_Train_scaled = scaler.fit_transform(feats_Train)
feats_Test_scaled = scaler.transform(feats_Test)

modelP = lm.Perceptron(random_state=42, max_iter=1000, tol=1e-3) # Add random_state and convergence params
modelP.fit(feats_Train_scaled, lbls_Train)
lbls_predP = modelP.predict(feats_Test_scaled)

print("\nAccuracy measures using Linear Model Perceptron:")
accP = m.accuracy_score(lbls_Test, lbls_predP)
results['Perceptron'] = accP
print(m.classification_report(lbls_Test, lbls_predP, target_names=['High', 'Low', 'Medium']))
print(f"\nAccuracy using Linear Model Perceptron: {accP:.3f}")

# --- 4. Logistic Regression ---
print("\n--- Training Logistic Regression ---")
t.sleep(0.5)
# Use scaled features
modelL = lm.LogisticRegression(random_state=42, max_iter=1000) # Add random_state and increase max_iter
modelL.fit(feats_Train_scaled, lbls_Train)
lbls_predL = modelL.predict(feats_Test_scaled)

print("\nAccuracy measures using Linear Model Logistic Regression:")
accL = m.accuracy_score(lbls_Test, lbls_predL)
results['Logistic Regression'] = accL
print(m.classification_report(lbls_Test, lbls_predL, target_names=['High', 'Low', 'Medium']))
print(f"\nAccuracy using Linear Model Logistic Regression: {accL:.3f}") # Corrected variable from accP to accL

# --- 5. MLP Classifier ---
print("\n--- Training MLP Classifier ---")
t.sleep(0.5)
# Use scaled features
# Increased max_iter, added early stopping for potentially better convergence/avoiding overfitting
modelN = nn.MLPClassifier(activation="logistic", # Sigmoid activation
                           hidden_layer_sizes=(100,), # Default: one hidden layer with 100 neurons
                           random_state=42,
                           max_iter=500, # Increased iterations
                           early_stopping=True, # Stop if validation score doesn't improve
                           n_iter_no_change=10)
modelN.fit(feats_Train_scaled, lbls_Train)
lbls_predN = modelN.predict(feats_Test_scaled)

print("\nAccuracy measures using MLP Classifier:")
accN = m.accuracy_score(lbls_Test, lbls_predN)
results['MLP Classifier'] = accN
print(m.classification_report(lbls_Test, lbls_predN, target_names=['High', 'Low', 'Medium']))
print(f"\nAccuracy using Neural Network MLP Classifier: {accN:.3f}")

# --- Summary of Accuracies ---
print("\n--- Model Accuracy Summary ---")
for model_name, accuracy in results.items():
    print(f"{model_name}: {accuracy:.3f}")
print("----------------------------")


# %% [markdown]
# ## Test Prediction with Specific Input
#
# Optionally, provide input values for the features ('raisedhands', 'VisITedResources', 'Discussion', 'StudentAbsenceDays') to get predictions from all trained models.

# %% [code]
# --- Test Specific Input ---

# Mapping from numerical prediction back to class labels (adjust if encoding was different)
# Based on the original script's final check: 0 -> H, 1 -> M, 2 -> L
# Let's verify this with the actual encoder if possible.
# Assuming 'Class' was encoded alphabetically: H (0), L (1), M (2)
# The original script used: if pred == 0: "H", elif pred == 1: "M", elif pred == 2: "L"
# Let's stick to the original script's mapping for consistency with its output.
pred_map = {0: "H", 1: "M", 2: "L"}

# Check if the 'StudentAbsenceDays' encoder exists and map input accordingly
absence_encoder = encoders.get('StudentAbsenceDays')

while True: # Loop until valid choice
    choice = input("\nDo you want to test specific input (y/n)? ").lower()
    if choice in ['y', 'n']:
        break
    else:
        print("Invalid input. Please enter 'y' or 'n'.")

if choice == "y":
    print("\nEnter values for the following features:")
    try:
        rai = int(input("Enter raised hands (numeric value): "))
        res = int(input("Enter Visited Resources (numeric value): "))
        dis = int(input("Enter no. of Discussions (numeric value): "))

        # Handle absence days input based on original categories
        while True:
            absc_input = input("Enter No. of Absences (Under-7 or Above-7): ").strip()
            if absc_input.lower() in ["under-7", "above-7"]:
                # Encode the input using the fitted encoder if available
                if absence_encoder:
                    # Find the encoded value for the input category
                    # Need to handle potential errors if input isn't in learned classes
                    try:
                       absc = absence_encoder.transform([absc_input])[0]
                       break
                    except ValueError:
                       print(f"Error: '{absc_input}' was not found during initial encoding. Available categories: {absence_encoder.classes_}")
                       # Fallback or re-prompt - Here we just inform and might error later
                       # For simplicity, let's try the original script's hardcoded logic if encoder fails
                       print("Attempting fallback encoding (Under-7: 1, Above-7: 0)")
                       if absc_input.lower() == "under-7":
                           absc = 1 # Original script mapping
                           break
                       else:
                           absc = 0 # Original script mapping
                           break

                else:
                    # Fallback if encoder wasn't stored (should not happen with current code)
                    print("Warning: Absence encoder not found. Using fallback: Under-7 -> 1, Above-7 -> 0")
                    if absc_input.lower() == "under-7":
                       absc = 1
                       break
                    else:
                       absc = 0
                       break
            else:
                print("Invalid input. Please enter 'Under-7' or 'Above-7'.")

        # Create the input array
        input_arr = np.array([[rai, res, dis, absc]]) # Reshape for single prediction

        # --- Make Predictions ---
        # Scale input for linear models and MLP
        input_arr_scaled = scaler.transform(input_arr) # Use the same scaler fitted on training data

        predD_num = modelD.predict(input_arr)[0]
        predR_num = modelR.predict(input_arr)[0]
        predP_num = modelP.predict(input_arr_scaled)[0]
        predL_num = modelL.predict(input_arr_scaled)[0]
        predN_num = modelN.predict(input_arr_scaled)[0]

        # Map numerical predictions to labels using the defined map
        predD = pred_map.get(predD_num, f"Unknown ({predD_num})")
        predR = pred_map.get(predR_num, f"Unknown ({predR_num})")
        predP = pred_map.get(predP_num, f"Unknown ({predP_num})")
        predL = pred_map.get(predL_num, f"Unknown ({predL_num})")
        predN = pred_map.get(predN_num, f"Unknown ({predN_num})")

        print("\n--- Predictions for Input ---")
        t.sleep(0.5)
        print(f"Using Decision Tree Classifier: {predD}")
        t.sleep(0.5)
        print(f"Using Random Forest Classifier: {predR}")
        t.sleep(0.5)
        print(f"Using Linear Model Perceptron: {predP}")
        t.sleep(0.5)
        print(f"Using Linear Model Logistic Regression: {predL}")
        t.sleep(0.5)
        print(f"Using Neural Network MLP Classifier: {predN}")
        print("\n-----------------------------")

    except ValueError:
        print("\nInvalid numerical input. Please enter numbers for raised hands, resources, and discussions.")
    except Exception as e:
        print(f"\nAn error occurred during prediction: {e}")

print("\nExiting...")
t.sleep(0.5)

# %% [markdown]
# --- End of Notebook ---
```
